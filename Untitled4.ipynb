# **2-checking missing values and qutliers**

```
#cheak for the null values 
insurance_df.isna().apply(pd.value_counts)
```

```
#cheching for outliers 
#Outliers

#Outliers are values that are very large or very small with respect to the distribution of the other data. 
#We can only find outliers in numerical data. Box plots are one good way to find the outliers in a dataset.

plt.figure(figsize=(20,15))

plt.subplot(3,1,1)
sns.boxplot(x= insurance_df.Inclination, color='red')

plt.subplot(3,1,2)
sns.boxplot(x= insurance_df.Name, color='red')

plt.subplot(3,1,3)
sns.boxplot(x= insurance_df.Eccentricity, color='red')

plt.show()


```

```
#visulizing data to make anlysis 
plt.figure(figsize=(20,15))
x= insurance_df.Eccentricity.index
y= insurance_df.Inclination

plt.xlabel('Eccentricity')
plt.ylabel('Inclination')
plt.bar(x,y,align='center',color='red' , edgecolor=('black'))

plt.show()
```
